```python
grid = {
    "Room A": "Dirty",
    "Room B": "Clean",
    "Room C": "Dirty",
    "Room D": "Dirty",
}

adjacency = {
    "Room A": ["Room B", "Room C"],
    "Room B": ["Room A", "Room D"],
    "Room C": ["Room A", "Room D"],
    "Room D": ["Room B", "Room C"],
}

def utility_based_cleaning_agent(grid, start_room):
    total_utility = 0
    internal_model = grid.copy()
    cleaned_rooms = []
    visited_rooms = []
    current_room = start_room

    while any(status == "Dirty" for status in internal_model.values()):
        print(f"\nüìç Current Room: {current_room}")
        print(f"üîç Internal Model: {internal_model}")
        print(f"üßº Cleaned Rooms: {cleaned_rooms}")
        print(f"üö∂‚Äç‚ôÇÔ∏è Visited Rooms: {visited_rooms}")
        print(f"‚ö° Total Utility: {total_utility}")

        # Cleaning the room
        if internal_model[current_room] == "Dirty":
            print(f"üßπ Cleaning {current_room}")
            internal_model[current_room] = "Clean"
            cleaned_rooms.append(current_room)
            total_utility += 10  # Reward for cleaning
        
        # Check revisit penalty
        if current_room in visited_rooms:
            print(f"‚ö†Ô∏è Penalty for revisiting {current_room} (-5)")
            total_utility -= 5

        visited_rooms.append(current_room)  # Track visited rooms

        # Find next dirty room
        next_room = None
        for neighbor in adjacency[current_room]:
            if internal_model[neighbor] == "Dirty":
                next_room = neighbor
                break
        
        # Move to the next room
        if next_room:
            print(f"‚û°Ô∏è Moving to {next_room} (-1 utility)")
            total_utility -= 1  # Cost of moving
            current_room = next_room
        else:
            break  # Exit when no dirty rooms remain

    print(f"\n‚úÖ Cleaning complete! üèÜ Final Utility: {total_utility}")

# Run the agent
utility_based_cleaning_agent(grid, "Room A")
```
