```python
grid = {
    "Room A": "Dirty",
    "Room B": "Clean",
    "Room C": "Dirty",
    "Room D": "Dirty",
}

adjacency = {
    "Room A": ["Room B", "Room C"],
    "Room B": ["Room A", "Room D"],
    "Room C": ["Room A", "Room D"],
    "Room D": ["Room B", "Room C"],
}


def utility_based_cleaning_agent(grid, start_room):
    total_utility = 0
    internal_model = grid.copy()
    cleaned_rooms = []
    visited_rooms = []
    current_room = start_room

    while any(status == "Dirty" for status in internal_model.values()):
        print(f"\nCurrent Room: {current_room}")
        print(f"Internal Model: {internal_model}")
        print(f"Cleaned Rooms: {cleaned_rooms}")
        print(f"Visited Rooms: {visited_rooms}")
        print(f"Total Utility: {total_utility}")

        # Cleaning the room
        if internal_model[current_room] == "Dirty":
            print(f"üßπ Cleaning {current_room}")
            internal_model[current_room] = "Clean"
            cleaned_rooms.append(current_room)
            total_utility += 10  # Reward for cleaning
        
        # Check revisit penalty
        if current_room in visited_rooms:
            print(f"Penalty for revisiting {current_room} (-5)")
            total_utility -= 5

        visited_rooms.append(current_room)  

        # Find next dirty room
        next_room = None
        for neighbor in adjacency[current_room]:
            if internal_model[neighbor] == "Dirty":
                next_room = neighbor
                break
        
        # Move to the next room
        if next_room:
            print(f"Moving to {next_room} (-1 utility)")
            total_utility -= 1  
            current_room = next_room
        else:
            break  

    print(f"\n‚úÖ Cleaning complete! üèÜ Final Utility: {total_utility}")

utility_based_cleaning_agent(grid, "Room A")
```
